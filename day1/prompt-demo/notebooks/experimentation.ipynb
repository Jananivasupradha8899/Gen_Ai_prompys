# 🧪 Prompt Engineering Experimentation Playground

Use this notebook to test and refine your prompts for **Chain of Thought**, **ReAct**, and **Self-Reflection**.

```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()  # Ensure you have a .env file with OPENAI_API_KEY
client = OpenAI()

def test_prompt(prompt_text, model="gpt-4o-mini"):
    response = client.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt_text}],
        temperature=0.7
    )
    return response.choices[0].message.content
```

### 1. Test Chain of Thought (CoT)

```python
query = "Can I return an item if I lost the box?"
cot_prompt = f"""
Customer Query: {query}

Helpful Rule: Items must be in original condition.

Think step-by-step before answering:
"""

print(test_prompt(cot_prompt))
```
